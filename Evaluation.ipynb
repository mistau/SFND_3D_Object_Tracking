```python
import numpy as np
import matplotlib.pyplot as plt

# get the data
xval = range(1, 18)
data = np.loadtxt('data/results.txt', usecols=range(1, 18))
num_rows, num_cols = data.shape
# print(data)

# get the labels
mylabel=[]
with open('data/results.txt') as infile:
    for line in infile:
        fields = line.strip().split()
        mylabel.append(fields[0])
#print(mylabel)
   
plt.figure(figsize=(13,8))
plt.title("Time To Collision from Camera")
plt.xlabel('frame#')
plt.ylabel('TTC [sec]')

i=0
for row in data:
    plt.plot(xval, row, label=mylabel[i])
    i = i+1

plt.legend(bbox_to_anchor=(1.0, 1.0), ncol=3)
# plt.ylim(0, 100)

plt.savefig('report/diagram.png')
```
